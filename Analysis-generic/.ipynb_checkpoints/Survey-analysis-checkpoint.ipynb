# Analysis of survey data for "Characterizing ORCID adoption in academic communities"
This notebook describes an analysis pipeline which is suitable for processing the data export from Qualtrics. 

### Functions and packages

```python
import pandas as pd, numpy, collections, matplotlib.pyplot as plt;

# some functions 
def groupMerge (col):
    grouped = data.groupby([col])[col].count().fillna(0)
    grouped = grouped.to_frame()
    codes = optionCodes[optionCodes.QID == col]
    grouped["option"] = grouped.index.astype(int)
    grouped = pd.merge(left=grouped, right=codes, left_on="option", right_on='option')
    grouped.index = grouped["optionText"]
    return grouped[grouped.columns[:-3]]

def twoGroupMerge (col0, col1):
    grouped = data.groupby([col0, col1])[col0].count().unstack(col1).fillna(0)
    #grouped = grouped.to_frame()
    codes = optionCodes[optionCodes.QID == col0]
    grouped["option"] = grouped.index.astype(int)
    grouped = pd.merge(left=grouped, right=codes, left_on="option", right_on='option')
    grouped.index = grouped["optionText"]
    return grouped[grouped.columns[:-3]]

def plotMCQ(colRegEx, matrix = None):
    """colRegEx should be a regular expression of the form r'Q2.\5_[0-9]*$'"""
    name = data.filter(regex=colRegEx, axis=1).melt()
    name["value"] = name["value"].fillna(0).astype(int)
    name = pd.merge(left=name, right=questionCodes, left_on='variable', right_on='QualtricsID')
    if matrix:
        name = pd.merge(left=name, right=optionCodes[optionCodes.QID==name.QID[0]], how="left", left_on="value", right_on='option') 
        name = name.groupby(["subQtext", "optionText"])["subQtext"].count().unstack("optionText").fillna(0)
        plot = name.plot(kind="barh", stacked=True)
        plot.legend(bbox_to_anchor=(1, 1))
        plot.set_ylabel("")
    else :
        name = name.groupby(["subQtext"])["value"].sum().fillna(0).to_frame()
        plot = name.sort_values("value").plot(kind='barh', legend=None, rot=0).set_ylabel("")
    return plot
```

```python
#read the data
data = pd.read_csv("ORCIDstudy_QualtricsSurvey_generic-v2_20180420.csv", quotechar='"')
data = data.iloc[:, 11:]

questionCodes = pd.read_csv("surveyQuestions.csv", quotechar='"')

optionCodes = pd.read_csv("Options.csv", quotechar='"')
```

```python
# create a dataframe with all questions and options based on the Qualtrics export
tdata = data.iloc[:1,:].transpose()
surveyQuestions = tdata.rename(columns={0: "QualtricsText"})
#tdata["QualtricsText"] = tdata["QualtricsText"].astype(str)
surveyQuestions["QID"] = surveyQuestions.index.str.extract("([Q]\d\.\d{0,2})", expand=False)
surveyQuestions["Qtext"] = surveyQuestions["QualtricsText"].str.extract("(.*?)-", expand=False)
surveyQuestions["subQID"] = surveyQuestions.index.str.extract("_(.*)", expand=False)
surveyQuestions["subQtext"] = surveyQuestions["QualtricsText"].str.extract("(?s:.*)-(.*)", expand=False)
surveyQuestions.index.names=["QialtricsID"]
surveyQuestions.to_csv("surveyQuestions.csv")
surveyQuestions
```

```python
#assign staff and student roles based on role selection
def studStaff (row):
    if row["Q6.2"]  in ("1", "2", "3","4"):
        return "staff"
    if row["Q6.2"] in ("5", "6"):
        return "student"
    if row["Q6.2"] == "7":
        return "other"
    return 

#group by affiliation and staff/students
data["staff or student"] = data.apply (lambda row: studStaff(row),axis=1)
affill = twoGroupMerge("Q6.1", "staff or student")

#plot the dataframe without the last 3 columns
affil[affil.columns[:-3]].plot(kind='barh', stacked=True).set_ylabel("")
#df2
affil
```
